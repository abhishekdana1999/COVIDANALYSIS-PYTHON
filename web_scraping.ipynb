```python
import pandas as pd
import json
import requests
```

```python
states = requests.get("https://api.livecovid.in/api/state")

stateparsed = json.loads(states.text)
```

```python
df = pd.DataFrame(stateparsed["statewise"])
```

```python
statename = []
for i in range(len(df.iloc[['1']].values[0])):
    statename.append(df.iloc[['1']].values[0][i])
```

```python
activecases = []
for i in range(len(df.iloc[['2']].values[0])):
    activecases.append(df.iloc[['2']].values[0][i])

```

```python
confirmedcases = []
for i in range(len(df.iloc[['3']].values[0])):
    confirmedcases.append(df.iloc[['3']].values[0][i])

```

```python
deathcases = []
for i in range(len(df.iloc[['4']].values[0])):
    deathcases.append(df.iloc[['4']].values[0][i])

```

```python
recovercases = []
for i in range(len(df.iloc[['6']].values[0])):
    recovercases.append(df.iloc[['6']].values[0][i])

```

```python
dates = []
for i in range(len(df.iloc[['7']].values[0])):
    dates.append(df.iloc[['7']].values[0][i])

```

```python
newData = pd.DataFrame()
```

```python
newData["States"] = statename
```

```python
newData["Active"] = activecases
```

```python
newData["Confirmed"] = confirmedcases
```

```python
newData["Deaths"] = deathcases
```

```python
newData["Recovered"] = recovercases
```

```python
newData["Date"] = dates
```
